## 라이브러리

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset

import gzip
import numpy as np
import pandas as pd
from sklearn import datasets, model_selection

from matplotlib import pyplot as plt
from matplotlib import cm
%matplotlib inline

import warnings
warnings.filterwarnings(action='ignore')
```

## 데이터 불러오기 & 탐색

```python
## f-mnist 데이터 불러오기
with gzip.open('data/f-mnist/train-images-idx3-ubyte.gz', 'rb') as f:
    mnist_data = np.frombuffer(f.read(), np.uint8, offset=16)
    print(mnist_data.shape)
    mnist_data = mnist_data.reshape(-1, 784) # 긴 벡터를 이미지 단위(28x28)로 쪼개기
    print(mnist_data.shape)
    
with pd.option_context('display.max_rows', 2, 'display.max_columns', None): # row는 2개만, column은 전부 출력      
    display(pd.DataFrame(mnist_data))
```

```python
mnist_data = mnist_data / 255 # 정규화
with pd.option_context('display.max_rows', 2, 'display.max_columns', None):      
    display(pd.DataFrame(mnist_data))
```

```python
## 첫번쨰 이미지 확인
first_image = np.copy(mnist_data[0].reshape(28, 28))
# np.around: decimals만큼 0.5를 기준으로 올림 혹은 내림, if decimals=1: 소수점 첫째자리까지
# display(np.around(first_image, decimals=1)) 

with pd.option_context('display.max_rows', None, 'display.max_columns', None):         
    display(pd.DataFrame(np.around(first_image, decimals=3))) # 첫번째 이미지 - 배열 형태

plt.imshow(first_image, cmap=cm.gray_r) # 첫번째 이미지 - 사진 형태
plt.show() 
```

```python
## 타겟변수 불러오기
with gzip.open('data/f-mnist/train-labels-idx1-ubyte.gz', 'rb') as f:
    mnist_label = np.frombuffer(f.read(), np.uint8, offset=8)
```

```python
print(mnist_label) # 전체 라벨
print(mnist_label[0]) # 첫번째 라벨 클래스
# 9는 순서상 Ankle Boot임
# T-shirt/Top, Trouser, Pullover, Dress, Coat, Sandal, Shirt, Sneaker, Bag, Ankle Boot 
```

```python
train_size = 5000 # 데이터를 일부만 사용
test_size = 500

train_X, test_X, train_Y, test_Y = model_selection.train_test_split(
    mnist_data, mnist_label, train_size=train_size,test_size=test_size)
```

```python
train_X = train_X.reshape((len(train_X), 1, 28, 28))
test_X = test_X.reshape((len(test_X), 1, 28, 28))
```

```python
print('shape is:', train_X.shape)
# 1은 흑백이 아닌 컬러 이미지일 때를 위한 (지금은 없지만) 추가 차원이다.

print('first train set image in array:')
print('shape is:', train_X[0].shape)
with np.printoptions(threshold=np.inf):
    print(train_X[0])  # 첫번째 train set 이미지 - 배열 형태
```

```python
## numpy -> 텐서 변환
train_X = torch.from_numpy(train_X).float()
train_Y = torch.from_numpy(train_Y).long()

test_X = torch.from_numpy(test_X).float()
test_Y = torch.from_numpy(test_Y).long()

print(train_Y.shape)
print(test_X.shape)
print(test_Y.shape)
```

```python
train = TensorDataset(train_X, train_Y) # 피쳐변수와 타겟변수 텐서를 합침

print(train[0]) # 텐서의 첫 번째 데이터 역시 확인 가능
# 맨 끝에 tensor(1) 있음

train_loader = DataLoader(train, batch_size=100, shuffle=True) # 미니배치 분할
```

## 모델 학습 및 선정

```python
# class Net(nn.Module):
#     def __init__(self):
#         super(Net, self).__init__()
#         # 합성곱층
#         self.conv1 = nn.Conv2d(1, 6, 5) # 입력 채널 수, 출력 채널 수, 필터 크기
#         self.conv2 = nn.Conv2d(6, 16, 5)
#         # 전결합층
#         self.fc1 = nn.Linear(256, 64) # 256 = (((28-5+1)/2 )-5+1)/2 * (((28-5+1)/2 )-5+1)/2 * 16
#         self.fc2 = nn.Linear(64, 10)
    
#     def forward(self, x):
#         x = F.max_pool2d(F.relu(self.conv1(x)), 2) # 풀링 영역 크기
#         x = F.max_pool2d(F.relu(self.conv2(x)), 2)
#         x = x.view(-1, 256)
#         x = F.relu(self.fc1(x))
#         x = self.fc2(x)
#         return F.log_softmax(x) # softmax 통해서 확률 형태로 값 배출
    
# model = Net()
```

```python
# sequential
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.architecture = nn.Sequential(
            nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=0),
            nn.ReLU(True),
            nn.MaxPool2d(kernel_size=2),
                        
            nn.Conv2d(6, 16, 5),  # 입력 채널 수, 출력 채널 수, 필터 크기
            nn.ReLU(True),
            nn.MaxPool2d(2), # 필터 크기
            
            nn.Flatten(),
            
            nn.Linear(256, 64),
            nn.ReLU(True),
            
            nn.Linear(64, 10),
            nn.ReLU(True)
        )
        
    def forward(self, x):
        x = self.architecture(x)
        return F.log_softmax(x)
    
model = Net()
```

```python
criterion = nn.CrossEntropyLoss() # 오차함수 객체

optimizer = optim.SGD(model.parameters(), lr=0.01) # 최적화 객체
```

```python
# 학습
for epoch in range(50): # 50회
    total_loss = 0
    for train_x, train_y in train_loader: # 분할해 둔 데이터 사용
        optimizer.zero_grad() # gradient 초기화
        output = model(train_x) # 순전파
        loss = criterion(output, train_y) # loss 계산
        loss.backward() # 역전파 계산
        optimizer.step() # 가중치 weight 업데이트 
        total_loss += loss.item() # 누적오차 계산, loss.item(): loss의 스칼라값
    # 5회 반복마다 누적 오차 출력
    if (epoch+1) % 5 == 0:
        print('epoch:', epoch+1, ' total_loss:', total_loss)
```

```python
result = torch.max(model(test_X).data, 1)[1] # torch.max(): 주어진 텐서 배열의 최대 값이 들어있는 index를 리턴하는 함수
accuracy = sum(test_Y.data.numpy() == result.numpy()) / len(test_Y.data.numpy()) # 모형의 정확도 측정

print('Accuracy:', accuracy*100, '%')
# random baseline: 10%
```
