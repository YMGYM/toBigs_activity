## 데이터 전처리

앞서 분석한 결과를 가지고 데이터 전처리를 진행했다.

```python
processed_data = data.copy() # 수정을 위해 데이터프레임을 복사한다.
```

```python
processed_data = processed_data.drop(['Unnamed: 0','User_ID','Product_ID','Product_Category_2', 'Product_Category_3'], axis=1) # 중분류 소분류를 제거해 주었다.
```

```python
target = processed_data['Gender'] # Target 분리
processed_data = processed_data.drop(['Gender'], axis=1)
```

```python
processed_data # 피처가 제거된 후의 데이터프레임 표시
```

KNN 분류를 하기 위해서는 Input이 numerical 해야 하므로 ont-hot 형태로 변환해줄 필요가 있다.

따라서 범주형 변수에 대해 변환을 진행해 주었다.

```python
age_dummy = pd.get_dummies(data=processed_data['Age'], prefix="age")
occupation_dummy = pd.get_dummies(data=processed_data['Occupation'], prefix="occupation")
city_dummy = pd.get_dummies(data=processed_data['City_Category'], prefix="city")
stay_dummy = pd.get_dummies(data=processed_data['Stay_In_Current_City_Years'], prefix="stay")
marital_dummy = pd.get_dummies(data=processed_data['Marital_Status'], prefix="marital")
product_dummy = pd.get_dummies(data=processed_data['Product_Category_1'], prefix="product")
```

`Purchase` 값도 0과 1 사이의 값으로 정규화를 진행했다.

one-hot이 0과 1로 이루어졌기 때문에 단위를 통일해 주면 좋을 것 같다고 생각했다.

```python
maximum, minimum = processed_data["Purchase"].max(), processed_data["Purchase"].min()
purchase_scale = pd.DataFrame([(x - minimum) / (maximum - minimum) for x in processed_data["Purchase"]], columns=['purchase'])
```

```python
# 위에서 생성한 값들을 전부 하나의 데이터프레임으로 만들었다.
train_data = pd.concat([age_dummy, occupation_dummy, city_dummy, stay_dummy, marital_dummy, product_dummy, purchase_scale ], axis=1)
```

```python
train_data
```

마찬가지로 `target` 에도 `'M'`, `'F'` 로 된 값을 수치값인 0과 1로 바꿔주는 작업을 취했다.
익명함수를 생성하고, `apply` 함수를 통해 시도했다.

```python
# target의 M과 F를 0과 1로 변경하는 익명함수 생성
change_gender = (lambda x: 0 if x == 'M' else 1)
changed_target = target.apply(change_gender)
```

```python
changed_target
```

# KNN

위에서 구한 `train_data` 를 `train`, `test` 로 나눠서 학습했다.

처음에는 하이퍼파라미터 수정 없이 default 파라미터로 진행했다.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

점수 측정을 위한 `x_test`, `y_test` 변수를 만들었다.

비율은 20%로 분할했다.

```python
x_train, x_test, y_train, y_test = train_test_split(train_data, changed_target, random_state=0, test_size=0.20)
```

```python
# 데이터 shape 표시
print(f"x_train.shape : {x_train.shape} x_test.shape : {x_test.shape} y_train.shape : {y_train.shape} y_train.shape : {y_train.shape}")
```

### KNN 모델 생성 후 평가

KNN모델은 새로운 데이터를 모델에 넣고, 근처에 있는 데이터들의 군집을 참고하여 Classification을 진행한다.

sklearn의 기본 파라미터로 모델을 생성 후 평가해 보았다.

기본값은 다음과 같다
- K-neighbors : 5개
- weights : 가중치 없이(동일하게)
- 거리 계산 : 유클리디언 (p = 2, metric="minkowski")

```python
knn_model = KNeighborsClassifier()
```

```python
knn_model.fit(x_train, y_train) # 피팅
```

모델의 정확도(mean accuracy) 와 crosstab (Confusion Matrix) 을 측정해 보았다.

```python
train_score = knn_model.score(x_test,y_test)
print(f"test score : {train_score}")
```

```python
result = knn_model.predict(x_test)
```

```python
pd.crosstab(y_test, result,  rownames=["Actual"], colnames=["Prediction"])
```

```python
from sklearn.metrics import classification_report # 전체 보고서 report
```

```python
print(classification_report(y_test, result, target_names = ["Male", "Female"]))
```

classification report 를 출력해 보았다.

- `precision` 이란 모델이 예측한 것 중에서 실제로 정답인 비율이다.
- `recall` 이란 전체 데이터 중 모델이 정답을 맞춘 비율이다.
- `f1-score` 는 `precision` 과 `recall` 의 조화평균이다. (큰 비중이 가지는 bias 를 줄이기 위해서 사용한다.) 

전반적으로 `Female` 피처에 대해 정확도가 낮은 것을 확인할 수 있다. 이는 데이터의 성비가 균등하지 않아 생기는 문제로 보인다.

### 거리 가중치 모델 (손튜닝)
거리를 계산할 때 가까이 있는 데이터 간 가중치를 주는 모델을 생성해 데이터를 비교해 보았다.

```python
# 거리의 역수로 가중치를 주는 모델 생성
knn_weight = KNeighborsClassifier(weights="distance")
```

```python
knn_weight.fit(x_train, y_train)
```

```python
# 평가점수
train_score = knn_weight.score(x_test,y_test)
print(f"test score : {train_score}")
result = knn_weight.predict(x_test)
```

```python
pd.crosstab(y_test, result,  rownames=["Actual"], colnames=["Prediction"])
```

```python
print(classification_report(y_test, result, target_names = ["Male", "Female"]))
```

평가점수가 약간 상승한 것을 확인할 수 있었다.

## Grid Search

Grid Search 방법을 통해 하이퍼파라미터 튜닝을 시도했다.

Grid Search 는 하이퍼파라미터의 풀을 설정한 뒤에 가능한 모든 가능성을 전부 대입해서 결과를 비교해 보는 방법이다.

튜닝이 필요한 파라미터는 다음과 같았다.
- `n_neighbors` : 이웃한 점의 개수(K). 분류해야할 클래스가 2개 이므로, 홀수 개수로 시도해 보겠다.
- `metric` : 거리 계산 방식. 수업 시간에 다룬 3가지 메트릭스('euclidean', 'manhattan', 'mahalanobis') 를 시도해 보았다.
    - 추가 : 'mahalanobis' 는 covariance 를 인수로 넣어 줄 필요가 있었다.
    - 추가2 : 'mahalanobis' 계산 시 k-fold 에서 NaN 값이 발생하는 문제가 있어서(pow() 함수에서 자리수 문제로 추정됨) 제거 후 진행했다.
- `weight` : 거리의 역수로 가중치가 들어갈지 판단. True와 False 둘다 시도해 보았다.

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_params = {
    'n_neighbors': [i for i in range(1, 20) if i % 2 == 1], # 1부터 20까지 홀수만 생성
    'metric': ['euclidean', 'manhattan'],
    'weights': ['uniform', 'distance']
}
```

```python
grid_params
```

`GridSearchCV` 에는 이미 cross-validation이 포함되어있다.

따라서 "K-fold 교차 검증"은 따로 진행할 필요가 없을 것으로 생각된다.

```python
grid_model = GridSearchCV(
    estimator = KNeighborsClassifier({'VI' :np.linalg.inv(np.cov(x_train))}), # Mahalanobis 방법에서 사용하기 위한 파라미터
    param_grid = grid_params,
    scoring = 'accuracy', # 정확도로 파악
    n_jobs=-1, # -1 인경우 모든 cpu 코어를 다 사용해서 계산한다(계산 속도가 빨라진다)
    cv = 5, # 5-fold cross validation을 사용하겠다.
    verbose=3, # 계산 시간과 점수(정확도) 를 표시한다.
    return_train_score = True
)
```

```python
grid_model.fit(x_train, y_train)
```

### 튜닝 결과

```python
grid_model.cv_results_ # 결과 반환 (딕셔너리 형태로 반환됨)
```

```python
grid_model.best_params_ # 가장 결과가 좋았던 파라미터 반환
```

`euclidean` 방식에 `15` 개의 군집, 거리별로 가중치를 두는 조건의 경우 가장 점수가 높았던 것을 확인할 수 있었다.

하지만 마할라노비스 거리 계산은 포함되지 않았기 때문에, 이를 계산시 값이 달라질 수도 있을 것이다.

## 정리
- 데이터 분석 결과 큰 상관관계가 발생하는 것은 아니었다.
- 각 데이터들을 one-hot 형태로 변경한 뒤에  KNN모델로 학습했다.

- 그리드 서치를 통한 하이퍼파라미터 튜닝을 진행했다. 판단 기준은 정확도
- 튜닝 결과 남성에 대한 f1-score가 증가했으나 여성에 대한 f1-score가 소폭 하락한 것을 볼 수 있었다
- 하지만 데이터셋 특징으로 볼 때 성별 별 정확도 보다는 전반적인 정확도가 더 중요할 것으로 판단되어 이대로 두는 것이 좋을 것 같다.
