## Embedding

- 수업에서 다룬 임베딩 방법에는 One-hot encoding, CBOW, Skip-gram, GloVe, FastText가 있었습니다. 다양한 시도와 '비교' 결과를 함께 적어주세요! 파라미터를 조정해가는 과정도 해석에 도움이 될 수 있겠죠 :)

```python
# 기본적인 Vocab 만들기
def get_vocab(text):
    word2idx = {}
    idx2word = {}
    index = 0
    
    for sent in text:
        for t in sent:
            if word2idx.get(t) == None: # 보카브러리에서 단어가 가지고 와 지지 않는다면
                word2idx[t] = index
                idx2word[index] = t
                index += 1

    return idx2word, word2idx
            
```

```python
idx2word, word2idx = get_vocab(spam['wt'])
```

```python
word2idx['free'] # free 의 index는 508번이다.
```

### One-hot Encoding

One-Hot 인코딩은 데이터가 해당되는 index에 1을 저장하고, 나머지는 모두 0으로 만든다.

가장 간편한 방법 중 하나이나, 단어들 간 관계성을 설명하기는 어렵다는 단점이 있다.

즉 아래에서 나오는 평가는 할 수 없다.

```python
import numpy as np
def get_onehot(idx2word):
    l = len(idx2word)
    result = np.identity(l)
    # n * n 단위행렬을 생성한다. idx2word의 모든 단어들은 한 번씩만 나오는 단어임이 확실하므로, 단위 행렬을 만든 뒤, 인덱스를 불러오는 방식으로 만든다.
    
    return result
```

```python
onehot_wt = get_onehot(idx2word)
```

```python
# free의 one-hot벡터
onehot_wt[word2idx['free']]
```

### CBOW Encoding

`CBOW`은 한 단어가 나오는 이유에는 주위 단어들이 영향을 준 다는 가정에서 나오는 인코딩 방식이다.

주위 단어를 통해 가운데 단어를 예측하는 인공신경망을 만든 뒤에 가중치를 추출하는 방법으로 구한다.

```python
from gensim.models import Word2Vec, FastText
```

### FastText

FastText는 페이스북 연구팀에서 만든 Word Embedding 방법이다.

Word2Vec과 기본 원리는 동일하지만 부분 단어만 임베딩하는 방법을 사용한다.

```python
ftModel = FastText(sentences=spam['lemt'], window=5, min_count=1, workers=4, iter=2000, sg=1) # 스킵-그램 방식으로 학습하겠음
```

```python
ftVectors = ftModel.wv.vectors
ftI2W = ftModel.wv.index2word

ftVectors.shape
```

### Negative Sampling
Word2Vec 방법과 학습 방법은 비슷하지만, 단어가 나올 확률로서 비교하는 방식.

나오는 단어를 맞추는 신경망에서 이진 분류를 하는 신경망으로 결정하는 방식이다.

```python
n_ftModel = FastText(sentences=spam['lemt'], window=5, min_count=1, workers=4, iter=2000, sg=1, negative=10) # negative가 0보다 크면 negative sampling 사용.
```

```python
n_ftVectors = n_ftModel.wv.vectors
n_ftI2W = n_ftModel.wv.index2word

n_ftVectors.shape
```

### Wordcloud

Wordcloud를 사용하여 각 단어가 사용된 빈도를 시각화해서 보여줄 수 있다.

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(width=5000, height=3000, background_color="white", random_state=0)
plt.imshow(wc.generate(' '.join(spam['lower'])))
plt.axis("off")
plt.show()
```

WordCloud를 그려 보니 전체 데이터에서 now, u, will, call 등이 많이 나타났다.

메일 데이터임을 고려해 볼 때, 추후 다른 연락을 기다리겠다는 내용을 많이 포함할 것 같다는 생각이 든다.

```python
# Spam 메일인 단어들로만 그려진 wordcloud
wc = WordCloud(width=5000, height=3000, background_color="white", random_state=0)
plt.imshow(wc.generate(' '.join(spam[spam['v1']=='spam']['lower'])))
plt.axis("off")
plt.show()
```

스팸 메시지인 글들로만 WordCloud를 그려 보았다.

여기 들어가는 내용을 적절히 파악하면 스팸 메일임을 인지하는데 큰 도움이 될 수 있을 것으로 보인다.

주로 'free'가 자주 나왔으며, 'urgent', 'now'와 같이 시기를 강조하는 말이 자주 나온것을 확인할 수 있었다.

또한 'win', 'won'과 같이 사용자를 유혹할 수 있는 내용들이 있다고 생각이 든다.

```python
# ham 메일인 단어들로만 그려진 wordcloud
wc = WordCloud(width=5000, height=3000, background_color="white", random_state=0)
plt.imshow(wc.generate(' '.join(spam[spam['v1']=='ham']['lower'])))
plt.axis("off")
plt.show()
```

스팸이 아닌 메일들로 wordcloud를 다시 그려 보았다.

대부분의 단어가 spam메일의 단어들과 차이가 있는 것을 확인할 수 있었다.

'u'와 같은 단어는 공통적으로 등장하지만 (주어 등이므로) 대부분의 큰 단어들이 위 그림에서는 찾아보기 어려웠다.

앞서 만들었던 `show_similar` 함수를 사용해 스펨 메일의 위험성이 높은 단어들의 동의어를 파악해 보고자 한다.

```python
spam_words = ['free', 'call', 'text', 'now', 'txt', 'reply', 'mobile']

for w in spam_words:
    print(f"similar word of {w}")
    show_similar(w, 5)
    print('\n')
```

유사도가 높은 상위 5개의 단어들을 모델별로 출력했다.

유사도가 비슷한 단어를 활용해 'free' 라는 단어 대신 'club' 이라는 단어가 자주 보이거나,

'now' 대신 'today'라는 단어가 자주 보이는 등의 메일 또한 스팸 메일로 분류될 수 있을 것이라고 생각된다.
