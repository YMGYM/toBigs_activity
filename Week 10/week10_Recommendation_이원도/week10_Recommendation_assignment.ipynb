## 1. Content-based Filtering 실습

> content-based filtering은 movie(아이템) 간의 유사도를 계산하여 해당 movie와 유사한 movie를 추천해주는 간단한 추천 알고리즘
제공된 데이터에서 movie의 콘텐츠를 확인할 수 있는 데이터는 movies.csv의 장르와 tags.csv의 태그 정보가 있음

```python
# 아이템 간 유사도 계산을 위한 데이터프레임 만들기
# 열은 영화 장르가 될 수도 있고, 영화 태그가 될 수도 있음 (영화 태그의 경우 모든 영화에 대한 태그는 없음)
movies = pd.read_csv("ml-latest-small/movies.csv", index_col='movieId') # 일단 파일 불러와 보기
movies # index column을 movie의 id로 설정했다.
```

- DataFrame 을 확인해 본 결과, `title`과 `genres` 열을 내가 필요로 하는 것을 볼 수 있다. 
- `movieId` 행은 사용하지 않으므로 제거해 보겠다.
- `genres` 행은  '|' 을 기준으로 장르가 분류되어 있는 것을 확인할수 있었다.

```python
# 장르를 구하는 함수 생성. Series데이터를 모두 돌면서 |으로 나눈 뒤, 데이터가 존재하지 않으면 append 한다.
def get_genres(col):
    genres = []
    for rows in col:
        for g in rows.split('|'):
            if g not in genres:
                genres.append(g)
                
    return genres
            
```

```python
all_genres = get_genres(movies.genres)
all_genres # 잘 추출된 것을 확인할 수 있다.
```

Split 후 pivot 메소드를 사용하는 방법도 있었지만, 수동으로 구현해 보았습니다.

```python
piv = pd.DataFrame(columns = ["movieId"] + all_genres) # 계산을 위한 DataFrame 생성
piv
```

```python
for index, rows in movies.iterrows(): # movies 테이블을 열 단위로 반복한다.
    genresDict = {"movieId" : int(index)} # movieId 를 사요
    for g in rows.genres.split('|'):
        genresDict[g] = 1 # 해당되는 장르에서만 1값을 가진다.
        
    piv = piv.append(pd.DataFrame([genresDict]), ignore_index=True)
piv
```

```python
piv.set_index('movieId', inplace=True) # movieId를 인덱스로 설정한다.
piv # 정상적으로 생성되었다.
```

"TF-IDF" 에 대한 개념이 부족한 것 같아서 조금 공부를 다시 진행했습니다.

- TF는 Term Frequency 이며, 특정 장르에 대해서 사용자 별 장르가 나온 빈도수를 확인합니다. (특정 사용자에게서 장르가 n번 등장하면 tf(d, t) = 1 방식으로 나타낸다.)
- IDF는 Inverse Document Frequency 이며 한 장르가 등장한 문서의 수를 역수를 취한 것을 나타냅니다. n개의 문장에서 df번 장르가 등장했다고 치면, (n/df)를 사용합니다. 하지만 이 부분에서 n이 커질 수록 값이 기하급수적으로 커지는 모습을 보이기 때문에, log를 취하며, divide by 0을 방지하기 위해 1을 분모에 더해줍니다.
- TF-IDF : TF * IDF이며, 각 행(문서)에 대해 IDF값을 곱해서 구합니다.
- 이를 통해 전체 사용자에게서 장르가 가지는 중요성을 파악할 수 있다. 한 장르가 많이 나오는 사용자가 있을 수도 있지만, 이 사용자가 전체에서 하나 뿐이라면.. 이 장르의 중요성이 낮게 설정되어야 하기 때문이다.

콘텐츠 기반 필터링에서 TF-IDF를 실행해야하는 이유는 '빈도'에 관한 정보이기 때문이라고 생각한다. TF-IDF는 더 많은 빈도에 대한 정보를 얻을 수 있었다.


```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies['corpus'] = movies['genres'].apply(lambda x : ' '.join(x.split('|'))) # genres를 잘라서 corpus별로 분류합니다.
```

```python
tfidv = TfidfVectorizer().fit(movies['corpus']) # tf-idf 를 구하기 위해 transformer를 적용합니다
```

```python
temp = pd.DataFrame(tfidv.transform(movies['corpus']).toarray()) # tfidv를 적용하여 저장합니다.
```

```python
temp # 정상적으로 동작하는 것을 확인할 수 있었습니다.
```

```python
# 아이템 간 유사도 구하기
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# 영화-영화의 유사도를 코사인 유사도를 사용해서 구해줍니당
CBF = cosine_similarity(temp) # 코사인 유사도를 출력합니다.
```

```python
# movie - movie 아이템 간 유사도 데이터프레임
CBF = pd.DataFrame(CBF)
CBF.index = piv.index 
CBF.columns = piv.index # 인덱스와 컬럼명을 같게 수정 후 출력합니다.
CBF
```

```python
# 각 영화와 다른 영화와의 유사도가 모두 계산되어 있고, 유사도가 높은 영화가 가장 비슷한 영화라고 할 수 있습니당
# 입력 아이템과 가장 유사한 아이템 TOP10개를 뽑아봅시당
def getCBF(movieId):
    print(f"입력 영화: {movies.loc[movieId]['title']}")
    print(f"입력 영화 장르: {movies.loc[movieId]['genres']}")
    
    sort_data = CBF[movieId].argsort()[::-1].drop(movieId)[:10] # CBF 테이블에서 해당하는 데이터를 뽑아온 뒤 유사도가 높은 10개의 데이터를 구합니다.
    

    print("=====Most similar List=====")
    df = pd.DataFrame(movies.loc[sort_data.index]) # df를 출력합니다
    return df
```

## 2. Collaborative Filtering 실습

**User-based** CF를 실습해 봅시당

```python
# User-Item 간 선호도 데이터프레임 만들기

rating = pd.read_csv('ml-latest-small/ratings.csv')
rating
```

```python
movID = rating.movieId.unique()
usrID = rating.userId.unique()
```

```python
ratingTable = pd.DataFrame(index = usrID, columns =movies['movieId'])
ratingTable.index.name="userId"
ratingTable # Unique()를 사용해서도 Table을 만들 수 있습니다. 기본 테이블을 만들고 데이터를 채워 넣어 보겠습니다.
```

```python
# 해당 데이터프레임은 사용자-영화 선호도 데이터프레임입니당
for idx, rows in rating.iterrows():
    if rows.userId in usrID and rows.movieId in movID: # 데이터가 ratingTable에 있다면.
        ratingTable.loc[int(rows.userId), int(rows.movieId)] = rows.rating
    else:
        print(rows.userId, rows.movieId)
    
ratingTable
```

코사인 유사도에 대한 결과물은 이미 하단에 작성되어 있어서 다른 상관계수를 사용해 보고자 했습니다.

[이 포스팅](https://bskyvision.com/116)을 참고해서 확인해 보았으며 'Spearman 상관계수' 를 사용했다.

스피어만 상관계수는 두 번수간 '단조성'을 확인하는 계수이다. 단조성이 높으면 한 변수의 값이 증가하면, 다른 변수의 값도 같이 증가한다는 의미이다.

식은 다음과 같다.



피어슨 상관계수가 아니라 스피어만 상관계수를 사용한 데는 두 영화가 선형적인 관계로 평점이 올라가지 않을 수도 있지 않을까 생각해서이다.

```python
# 사용자 간 유사도 구하기
from scipy.stats import spearmanr # 스피어만 유사도를 임포트했다.
```

```python
tmpTable = ratingTable.fillna(0)
```

```python
sp = spearmanr(tmpTable, axis=1) # row에 대해 상관계수를 구해야 한다.(협력 기반 필터링 기술)
```

```python
CF = pd.DataFrame(sp[0])
```

```python
# 사용자와 사용자의 유사도를 구할 때 꼭 코사인 유사도를 사용하지 않아도 됩니당
# 보완 코사인 유사도 혹은 피어슨 상관계수를 적용해보아유

# 사용자 ID로 행,열의 인덱스를 설정했습니다.
CF.index = usrID
CF.columns = usrID
CF
```

```python
user = 604
target1 = CF[user].argsort()[::-1].iloc[0]
target2 = CF[user].argsort()[::-1].iloc[1]
```

```python
userMean = ratingTable.loc[user].mean(skipna=True) # 추천 대상 사용자 선호도 평균
target1Mean = ratingTable.loc[target1].mean(skipna=True)
target1Mean = ratingTable.loc[target2].mean(skipna=True)
```

```python
# 입력 사용자와 가장 가까운 유저에서 선호도가 가장 높은 아이템 TOP5를 추천해봅시당!
# 이 때, 입력 사용자가 이미 시청한/평가한 영화는 제외해야하도록 추천 리스트를 만들어봅시당
import matplotlib.pyplot as plt
def getCF(userId):
    recommend = []
    target = CF.loc[userId].argsort()[::-1].iloc[0] # 가장 가까운 유저 선택
    movieRanking = ratingTable.loc[target].sort_values(na_position='first')[::-1] #본 영화 중 rating이 높은 순서대로 정렬
    score = ratingTable.loc[userId].isnull() # 확인하려는 유저가 본 영화인지 확인
    
    
    # 그래프 그리기 (점수가 높은 순서대로 정렬 후 막대 그래프로 표시)
    
    genreScore = piv[score].sum().sort_values()[::-1]
    genreScore.drop(labels='(no genres listed)', axis=0, inplace=True)

    plt.xticks(rotation='vertical')
    plt.title(f"User {userId}'s favorite genres")
    plt.bar(genreScore.index, genreScore.values)
    plt.show()
    
    for item in movieRanking.index: # 아까 정렬한 rating의 순서대로 인덱스 반환
        if score.loc[item] == False: # 사용자가 안 봤을 경우(score 테이블의 영화의 isNull이 False인 경우)
            recommend.append(item) # 아이템을 추가함
        if len(recommend) == 5: # 추천된 아이템 길이가 5개가 된 경우
            break # 반복문을 빠져나감
            
    print("=====Recommendatin List=====")
    return movies.loc[recommend] # movie 테이블을 사용해서 추천된 영화 목록을 반환함
```

```python
getCF(604)
```

## 3. Matrix Factorization 실습

같은 폴더에 있는 MF.py의 MatrixFactorization()을 적용해 봅시당

```python
import MF
```

```python
# 사용자-영화 선호도 데이터프레임입니당
ratingTable.head()
```

```python
# MF.py를 참고하여 인자를 입력해 보아유

# tmpTable은 앞서 상관계수를 구할 때 사용됐던 테이블이며, null 값을 0으로 변경한 테이블이다.
# K의 값은 50~300 의 수치 중 선택한다고 했는데, 컴퓨터의 성능 상 50으로 진행했다.
factorizer = MF.MatrixFactorization(tmpTable.to_numpy(), k=50, learning_rate=0.01, reg_param=0.01, epochs=300, verbose=True)
factorizer.fit()
R_reduced = factorizer.print_results()
```

```python
# Matrix Factorizaion으로 찾아낸 잠재요인을 기반으로 재생산한 행렬이네유
pd.DataFrame(R_reduced)
```

앞서 user-based 와 item-based 두 가지 방법으로 추천 리스트를 만들었지만, Matrix Fatorization을 사용해 보지는 못했다.

따라서 이를 사용해서 추천 리스트를 만들어 보고자 한다.

Matric Factorization을 사용한 추천의 장점은 다음과 같다.

- Cold Start problem을 겪지 않는다. (협업 필터링 전체의 장점)
- 정확도가 높다
- 유연성을 가진다.
- Sparse 한 행렬에도 적용할 수 있다.


[이 게시글](https://ekdud7667.tistory.com/entry/MF-MATRIX-FACTORIZATION-TECHNIQUES-FOR-RECOMMENDER-SYSTEMS) 을 참고해서 작성했다.


```python
# MF.py를 사용해 이미 잠재요인 행렬을 구해서 이를 사용하려 한다.
# 잠재요인 행렬을 사용해서 사용자가 보지 않은 영화들의 점수를 구하고, 이들을 크기순으로 정렬하여 보여주는 방식이다.

def mf_recommend(userId):
    userRecommend = pd.Series(R_reduced[(userId - 1)], index=movies.index) # 특정 사용자의 잠재요인 벡터 R_reduced의 인덱스는 0부터 시작하므로 1을 빼주었다.
    isNotWatched = ratingTable.loc[userId].isnull() # 확인하려는 사용자가 본 영화인지 확인 => 안봤으면 True
    rating = userRecommend[isNotWatched.values] # 사용자가 보지 않은 영화만 필터링해서 보여 준다.
    
    score = userRecommend.sort_values(na_position='first')[::-1] # 잠재요인 벡터를 정렬한 후 점수가 높은 것부터 정렬
    score = score[:10] # 10개로 잘라냈다.
    print("======== Recommendation List ===========")
    
    return movies.loc[score.index.values] # 잘라 낸 벡터를 기준으로 해당하는 영화를 보여줌
```
