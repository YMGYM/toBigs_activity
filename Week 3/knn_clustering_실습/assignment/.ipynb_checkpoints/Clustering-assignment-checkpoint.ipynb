## K-means Clustering

군집 중심(Centroid)를 설정하고, 데이터를 군집에 할당한 뒤에, 각 데이터들의 평균을 계산하여 정말로 중앙에 오도록 중심을 update하며 진행하는 방법

```python
from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters=4, random_state=0) # 클러스터 4개로 임시로 분류해 보겠다.
```

```python
kmeans_model.fit(noGender_data)
```

```python
label = kmeans_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = label, cmap='rainbow')
plt.show()
```

K-Means 알고리즘은 군집이 각각 구 형태이며, 비슷한 크기라고 가정한다.

따라서 원 모양으로 데이터가 군집화 된 것이 보인다.

### Elbow method

최적의 군집 개수를 찾기 위해 Elbow Method를 활용한다.
먼저 각각 군집을 확인하며, 그래프를 그려 보고자 한다.

```python
from scipy.spatial.distance import cdist

ks = [i for i in range(2, 10)]
wss = []

for k in ks:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(noGender_data)
    wss.append(sum(np.min(cdist(noGender_data, model.cluster_centers_, 'euclidean'),axis=1))) # 학습된 모델에서 각 점과 클러스터 중심간 거리 계산 (WSS)
```

```python
plt.title("WSS with Cluster number K")
plt.plot(ks, wss)
plt.xlabel("Cluster number")
plt.ylabel("WSS")
plt.show()
```

적절한 Elbow Point를 찾기 어렵다.

이는 군집의 개수가 뚜렷하지 않아서 생기는 문제로, 다른 평가방법을 생각해 내야 한다.

### Silhouette Coefficient

```python
from sklearn.metrics import silhouette_score # 실루엣 계수 계산하는 모듈 임포트
```

```python
ks = [i for i in range(2, 10)]
silhouette = []

for k in ks:
    model = KMeans(n_clusters=k, random_state=0)
    model.fit(noGender_data)
    silhouette.append(silhouette_score(noGender_data, model.labels_))
```

```python
plt.title("Silhouette with Cluster number K")
plt.plot(ks, silhouette)
plt.xlabel("Cluster number")
plt.ylabel("Silhouette Cofficient")
plt.show()
```

실루엣 계수를 확인해 보니, K = 9일 때 가장 높은 결과가 나오는 것을 확인할 수 있었다.

실루엣 계수는 -1~1 의 값을 가지지만, 최대값이 0.5가 나오지 못하는 것을 볼 때, 데이터가 적절하게 군집화가 되지 못한 것으로 생각할 수 있다.

K = 7일때도 두 번째로 높으며 9일때와 큰 차이가 없었다. 위의 Elbow Method에서도 K = 7에서 미약하게 그래프가 꺾이는 것을 확인할 수 있다.

따라서 모델의 간소화를 위해 K = 7로 놓고 시각화를 해 보고자 한다.

### K = 7 일때 시각화


```python
kmeans_model = KMeans(n_clusters=7, random_state=0) # 클러스터 4개로 임시로 분류해 보겠다.
kmeans_model.fit(noGender_data)
label = kmeans_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = label, cmap='rainbow')
plt.show()
```

군집화가 어느 정도 진행된 것을 확인할 수 있었다.

## DBSCAN

DBSCAN은 밀도 기반 군집화 방법이다.

정해진 거리 내에 점이 몇개 있는지에 따라서 Core Point, Border Point, Noise Point 로 나누고, 이를 기반으로 클러스터링한다.

```python
from sklearn.cluster import DBSCAN # DBSCAN 모델 불러오기
```

```python
dbscan_model = DBSCAN() # 기본 하이퍼파라미터로 실행했다. eps=5, min_samples = 5, metric= 'euclidean' 이다.
dbscan_model.fit(noGender_data)
```

```python
db_labels = dbscan_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = db_labels, cmap='rainbow')
plt.show()
```

```python
pd.DataFrame(db_labels).value_counts()
```

하나의 군집으로 묶여버린 것을 확인할 수 있었다.

이는 eps나, min_samples 의 문제로 보인다.

주 원인은 데이터가 0과 1사이의 값으로 정규화되었는데, eps가 0.5이기 때문으로 생각된다. 따라서 eps를 낮추어 다시 실행해 보았다.

```python
dbscan_model = DBSCAN(eps=0.1) # eps=0.1, min_samples = 5
dbscan_model.fit(noGender_data)
db_labels = dbscan_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = db_labels, cmap='rainbow')
plt.show()
```

```python
pd.DataFrame(db_labels).value_counts() # -1은 노이즈 데이터
```

조금이지만 군집화가 진행된 것을 확인할 수 있다. 하지만 노이즈 데이터인 -1이 매우 높은 비중을 차지하는 것을 확인할 수 있었다.

적절한 하이퍼파라미터 값을 찾기 위해 Elbow Method 를 사용해 보았다.

loss function 으로 `Nearestneighbors`를 통한 Elbow method 와 `Silhouette` 을 사용해 보았다.

```python
from sklearn.neighbors import NearestNeighbors
neigh_model = NearestNeighbors(n_neighbors=2)
neigh = neigh_model.fit(noGender_data)
distances, indices = neigh.kneighbors(noGender_data)
distances = np.sort(distances, axis=0)[:,1]
```

```python
plt.title("Neighbors with Cluster number K")
plt.plot(distances)
plt.xlabel("eps")
plt.ylabel("distance")
plt.show()
```

0.15보다 약간 작은 수 즈음에서 매우 가파르게 증가한 것을 확인할 수 있다.

```python
# 실루엣 계수
epses = np.arange(0.05, 0.2, 0.01)
silhouette = []

for eps in epses:
    model = DBSCAN(eps=eps)
    model.fit(noGender_data)
    silhouette.append(silhouette_score(noGender_data, model.labels_))
```

```python
plt.title("Silhouette with Cluster number K")
plt.plot(epses, silhouette)
plt.xlabel("eps")
plt.ylabel("Silhouette Cofficient")
plt.show()
```

0.2 또는 0.12에서 가장 높은 계수가 나온 것을 확인할 수 있다.

이는 Elbow Method 를 통해 유도한 것과 비슷한 결과를 보인다.
따라서 0.12를 채택했다.

```python
dbscan_model = DBSCAN(eps=0.12, min_samples=3) # eps=0.12, min_samples = 3 (임의의 수를 지정했다.)
dbscan_model.fit(noGender_data)
db_labels = dbscan_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = db_labels, cmap='rainbow')
plt.show()
```

```python
pd.DataFrame(db_labels).value_counts() # -1은 노이즈 데이터
```

아직까지 많은 데이터가(29개) noise 로 분류되었지만, 이전보다는 더 나은 분류 성능을 보인다.

### Hierarchical Clustering

계층적 군집화는 개체를 가까운 집단부터 묶어 나가며 클러스터링 하는 방식이다.

이 중 `Agglomerative` 방법은 Bottom-up 방식으로, 병합형 군집화이다.

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agglomer_model = AgglomerativeClustering(n_clusters=7) # 거리 측정 : ward, euclidean 클러스터 수는 k-means대로 7개로 설정
```

```python
agglomer_model.fit(noGender_data)
ag_labels = agglomer_model.labels_
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = ag_labels, cmap='rainbow')
plt.show()
```

```python
agglomer_model = AgglomerativeClustering(n_clusters=7, linkage='average')
agglomer_model.fit(noGender_data)
ag_labels = agglomer_model.labels_
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = ag_labels, cmap='rainbow')
plt.show()
```

다른 데이터 연결법을 사용해도 결과는 비슷하게 나타난다.

모델 성능 파악을 위해 실루엣 계수만 사용해서 비교해 보도록 하겠다.

```python
ks = [i for i in range(2, 10)]
silhouette = []

for k in ks:
    model = AgglomerativeClustering(n_clusters=k)
    model.fit(noGender_data)
    silhouette.append(silhouette_score(noGender_data, model.labels_))
```

```python
plt.title("Silhouette with Cluster number K")
plt.plot(ks, silhouette)
plt.xlabel("Cluster number")
plt.ylabel("Silhouette Cofficient")
plt.show()
```

7개에서 가장 높은 실루엣 계수가 나타났다.
따라서 7개의 군집은 적합하다고 볼 수 있을 것이다.

### 덴드로그램 생성

```python
from scipy.cluster.hierarchy import dendrogram, ward
```

```python
linkage_array = ward(noGender_data)
```

```python
dendrogram(linkage_array)
plt.xticks([]) # X좌표 지우기
plt.xlabel("samples")
plt.ylabel("cluster dist")

# 클러스터 개수를 구별하는 선 그리기
ax = plt.gca()
bounds = ax.get_xbound()
plt.plot(bounds, [2.7, 2.7], '--', c='k')
plt.plot(bounds, [4, 4], '--', c='k')
plt.plot(bounds, [1.5, 1.5], '--', c='k')
ax.text(bounds[1], 2.7, '4 cluster', va='center', fontdict={'size':15})
ax.text(bounds[1], 4, '2 cluster', va='center', fontdict={'size':15})
ax.text(bounds[1], 1.5, '7 cluster', va='center', fontdict={'size':15})
plt.show()
```

- bottom-up 방식의 병합 방식 과정을 덴드로그램을 통해 쉽게 시각화가 가능했다.
- 데이터 덴드로그램을 볼 때 클러스터 거리가 2.7 정도일때 4개의 군집이 생성된 것을 확인할 수 있었다.

```python
agglomer_model = AgglomerativeClustering(n_clusters=4, linkage='average')
agglomer_model.fit(noGender_data)
ag_labels = agglomer_model.labels_
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c = ag_labels, cmap='rainbow')
plt.show()
```

4개의 군집일 때 3d시각화를 해 보면 다음과 같은 모습이 나타났다.

## PCA 분석

주어진 데이터셋은 데이터 피처가 4개로 이루어져 있었고, 그래프로 표현 가능하게 만들기 위해 값이 두 개로 명확하게 갈라진 `Gender` 를 제거하고 진행했다.

`PCA` 는 행렬의 고유값(고유벡터) 를 사용해서 주어진 행렬에서 가장 핵심적인 차원 순서대로 남겨 놓은 뒤에 제거한다.

즉 PCA를 이용하면 차원을 축소시킬 수 있다.

위 군집화에서 가장 높은 점수를 보였던 K-means 모델을 사용해서 PCA된 데이터의 분석을 진행해 보고자 한다.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2) # 데이터 차원을 2개만 남기고 PCA 수행
pca.fit(scaled_data.T) # 데이터 Input
```

```python
pca_data = pca.components_.T # pca된 데이터 반환
pca_data.shape
```

```python
# 그래프에 표시해 보기
plt.scatter(x = pca_data[:,0], y = pca_data[:, 1])
plt.title("PCA processed data")
plt.show()
```

데이터가 3차원 보다 훨씬 분명하게 군집화가 된 것을 확인할 수 있었다.

'+' 모양 처럼 축이 나눠지고 있어서 클러스터를 4개로 분류해 보면 좋을 것 같다. 

```python
kmeans_model2 = KMeans(n_clusters=4, random_state=0) # 클러스터 4개로 분류해 보겠다.
kmeans_model2.fit(pca_data)
```

```python
label = kmeans_model2.labels_
plt.scatter(x = pca_data[:,0], y = pca_data[:, 1], c=label)
plt.title("K-means")
plt.show()
```

군집이 명확하게 잘 구분된 것을 확인할 수 있었다.

```python
silhouette_score(pca_data, kmeans_model2.labels_)
```

실루엣 계수로 0.6으로 상당히 올라간 것을 확인할 수 있었다.

다른 모델들로도 한번 수행해 보았다.

```python
dbscan_model2 = DBSCAN(eps=0.03)
dbscan_model2.fit(pca_data)
label = dbscan_model2.labels_
plt.scatter(x = pca_data[:,0], y = pca_data[:, 1], c=label)
plt.title("dbscan data")
plt.show()
```

```python
silhouette_score(pca_data, dbscan_model2.labels_)
```

DBSCAN 모델의 경우도 깔끔하게 분류가 되었으며, 3차원으로 분류한 0.13보다 더 높은 점수를 보인 것을 확인할 수 있었다.

```python
agglomer_model2 = AgglomerativeClustering(n_clusters=4)
agglomer_model2.fit(pca_data)
label = agglomer_model2.labels_
plt.scatter(x = pca_data[:,0], y = pca_data[:, 1], c=label)
plt.title("agglomerative data")
plt.show()
```

```python
silhouette_score(pca_data, agglomer_model2.labels_)
```

계층형 클러스터링을 사용해도 3차원 데이터로 클러스터링 한 것보다 높은 성과를 얻을 수 있었다.

## 정리

- 데이터 전처리 과정에서 특별한 상관관계를 보이지 않았다.
- 군집화를 위해 모든 데이터를 0과 1 사이로 정규화 한 뒤에 진행했다.

#### 3차원 데이터
- `Gender` 데이터는 0과 1 단 두개 밖에 가지지 않는 단순한 피처이기 때문에 3차원 시각화를 위해 제거하고 3개의 피처만 남겨 두었다.
- 클러스터링 결과 K-Means 방식을 사용한 모델이 가장 실루엣 계수가 높게 나왔으나, 잘 분류한다고 보기 어려웠다.

#### 2차원 데이터
- 데이터의 특징을 모두 살리면서 2차원 데이터로 시각하기 위해 PCA를 통해 차원을 축소했다.
- 차원을 축소한 데이터를 2차원 그래프에 그려 보니 3~4개의 클러스터로 구분이 가능했ㅏ.
- 3가지 모델을 사용해서 평가한 결과 K-Means 방식을 사용한 모델이 가장 실루엣 계수가 높았으며, 0.61로 3차원 때보다 높은 성능을 보였다
- 2차원 데이터는 하이퍼파라미터 튜닝을 진행하지 못해서 최고 성능이 나오는 모델이라고 하기 어려우나, 데이터 시각화의 직관적 해석으로 클러스터 4개가 가장 좋은 결과를 낼 것이라고 예상이 된다.
