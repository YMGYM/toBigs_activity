### 빠진 날짜 체크해 보기
측정이 안된 날이 있는지 체크해 보았다.

앞 시간에서 뒤 시간을 뺀 뒤에 1시간 간격이 아닌 위치가 있는지 체크해 보았다.

```python
import datetime
```

```python
missing_day = np.array([], dtype=int)
for i, x in enumerate(data["datetime"][1:]):
    if datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") - datetime.datetime.strptime(data["datetime"][i], "%Y-%m-%d %H:%M:%S") != datetime.timedelta(hours=1):
        missing_day = np.append(missing_day, i)
        
print(missing_day)
print("결측점 개수", len(missing_day))
```

## 히스토그램 및 이상치 분석
데이터의 이상치를 파악하기 위해 히스토그램을 그려 보았다.

```python
hist_data = ['temp', 'atemp', 'humidity', 'windspeed', 'casual', 'registered', 'count'] # 히스토그램을 그릴 데이터 산출
data[hist_data]
```

개인척인 추정으로 대여에 영향을 줄 수 있는 요인을 따로 추려 상관관계를 분석해 보았다

[피어슨 상관 계수](https://ko.wikipedia.org/wiki/%EC%83%81%EA%B4%80_%EB%B6%84%EC%84%9D) 에 의해
다음 규칙으로 생각한다.
- r이 -1.0과 -0.7 사이이면, 강한 음적 선형관계,
- r이 -0.7과 -0.3 사이이면, 뚜렷한 음적 선형관계,
- r이 -0.3과 -0.1 사이이면, 약한 음적 선형관계,
- r이 -0.1과 +0.1 사이이면, 거의 무시될 수 있는 선형관계,
- r이 +0.1과 +0.3 사이이면, 약한 양적 선형관계,
- r이 +0.3과 +0.7 사이이면, 뚜렷한 양적 선형관계,
- r이 +0.7과 +1.0 사이이면, 강한 양적 선형관계

여기서 `atemp` 와 `temp`,

`casual`, `registered` 와 `count` 는 강한 상관관계를 가지는 것을 확인할 수 있었다.

즉 서로가 서로의 값을 결정하는 데 큰 역할을 수행한다는 사실을 확인할 수 있다는 것이다.

따라서 `atemp` 와 `count` 만 남긴 채 다시 한번 상관관계를 그려 보았다.

```python
heat_data = ['atemp', 'humidity', 'windspeed', 'count']
sns.heatmap(data[heat_data].corr(),cmap = 'YlGnBu', annot=True) # annot를 사용해 숫자 표시.
plt.show()
```

각 기상 상태에 따른 총 대여수의 상관관계를 비교해 보았다.

자전거를 타는 주체는 사람이기 때문에, 사람이 느끼는 체감기온인 `atemp` 를 중요하게 생각해 남겨 놓았다.

상관계수를 확인해 보면 다음과 같은 결과가 나타났다.
- 기온과 대여량에는 양의 선형관계가 있다고 할 수 있었다.
- 습도와 대여량에는 음의 선형관계가 있다고 할 수 있었다.
- 풍속과는 큰 상관관계가 드러나지 않았다.

## Log Transformation

대여 수와 관련된 데이터가 왼쪽으로(낮은 값으로) 치우쳐저 있는 현상을 발견하여 로그 변환을 수행해 보았다

- `casual` , `registered` 의 경우, 0인 값이 존재하여 로그변환을 수행할 수 없었다.
- 모든 데이터에 1등을 더해 로그 변환을 수행할 수 있었지만, 두 데이터는 `count` 에 직접적으로 영향을 주는 데이터였다.
- 따라서 두 데이터는 사용하지 않을 것으로 생각해, 로그변환을 수행하지 않았다

```python
data["log_count"] = np.log(data["count"])
```

```python
f, axes = plt.subplots(ncols=2, figsize= (10,5))
sns.histplot(data["count"], ax = axes[0], color="g", kde=True)
sns.histplot(data["log_count"], ax = axes[1], color="g", kde=True) # 편향성이 줄어든다.
plt.show()
```

로그변환한 값을 사용하여 상관관계를 그려 보았다.

-  `atemp` 와 `log_count` 간의 상관관계가 0.37이므로, 약한 상관관계가 유지되고 있다.
- 또한 `humidity`와의 음의 상관계수도 약간 증가한 것을 확인했다.
- `windspeed` 와의 상관계수도 조금 증가했지만 아직까지는 유의미한 상관관계로 보기 어렵다

```python
heat_data = ['atemp', 'humidity', 'windspeed', 'log_count']
sns.heatmap(data[heat_data].corr(),cmap = 'YlGnBu', annot=True)
plt.show()
```

log 변환한 변수에 대해 pairplot은 다음과 갈다.

```python
sns.pairplot(data[heat_data])
plt.suptitle("Pairplot (로그변환 후)", size='xx-large')
plt.show()
```

## 범례별 자전거 대여

### 계절별
1은 봄이고, 2는 여름, 3은 가을 4는 겨울이다.
따라서 각각의 계절별로 따로 분류했다

```python
spring_count = data[data.season == 1]
summer_count = data[data.season == 2]
autumn_count = data[data.season == 3]
winter_count = data[data.season == 4]
```

```python
f, axes = plt.subplots(ncols=4, figsize=(20,5))
plt.suptitle("계절별 대여건수 (봄, 여름, 가을, 겨울)")
sns.histplot(spring_count["count"], ax=axes[0], bins=50)
sns.histplot(summer_count["count"], ax=axes[1], bins=50)
sns.histplot(autumn_count["count"], ax=axes[2], bins=50)
sns.histplot(winter_count["count"], ax=axes[3], bins=50)

plt.show()
```

대여건수가 적은 날도 많이 있기 때문에 위의 히스토그램으로는 한번에 보기가 어렵다.
따라서 대여건수가 30건 이하인 날을 따로 정리해서 다시 히스토그램을 그려 보았다.

```python
# 대여건수가 적은 상황에도 표시
f, axes = plt.subplots(ncols=4, figsize=(20,5))
plt.suptitle("계절별 대여건수 (봄, 여름, 가을, 겨울)")
sns.histplot(spring_count[spring_count["count"] < 30]["count"], ax=axes[0], bins=50)
sns.histplot(summer_count[summer_count["count"] < 30]["count"], ax=axes[1], bins=50)
sns.histplot(autumn_count[autumn_count["count"] < 30]["count"], ax=axes[2], bins=50)
sns.histplot(winter_count[winter_count["count"] < 30]["count"], ax=axes[3], bins=50)
plt.show()
```

확인 결과, 봄에는 다른 계절에 비해 대여건수가 적은 날 (3건 이하) 이 조금 더 많았다. 
반대로 가을에는 대여건수가 3건 이하인 날이 거의 적은 것을 확인할 수 있었다.

#### 계절별 평균 대여량
데이터를 계절별로 그룹화해 평균 대여량을 비교해 보았다.

```python
data_mean = data.groupby(["season"], as_index=False).mean()
```

```python
plt.bar(["봄", "여름", "가을", "겨울"], data_mean["count"])
plt.title("계절별 총 대여량")
plt.show()
```

계절 별로 데이터를 분석해 볼 때, 총 대여량은 가을 > 여름 > 겨울 > 봄 순으로 많았다.
또한 봄과 가을의 대여건수는 2배 가량의 큰 차이가 있음을 확인할 수 있었다.

```python
fig, axes = plt.subplots(ncols=2, sharey=True)
axes[0].bar(["봄", "여름", "가을", "겨울"], data_mean["casual"])
axes[0].set_title("계절별 비가입자 평균")

axes[1].bar(["봄", "여름", "가을", "겨울"], data_mean["registered"])
axes[1].set_title("계절별 가입자 평균")

plt.show()
```

전반적으로 비가입자의 수가 가입자 수에 비해 현저히 적은 모습을 나타낸다.

가입자의 경우는 겨울이 여름보다 살짝 많은 모습을 보였는데, 비가입자가 여름에 많이 유입되어 총합으로 볼 때는 여름이 겨울의 대여량보다 많아지게 되었다.

### 평일, 휴일별

`workingday` 를 기준으로 하여 따로 나눴다.

전체적인 사용량은 평일에 더 많이 나타나는 모습을 보였다.

즉 가입자는 통근 등을 위해 자전거를 대여하고, 비가입자는 휴일에 여가를 위해 많이 자전거를 대여했을 것으로 예측할 수 있다.

## 시간대에 따른 대여량 비교
자전거를 대여하는 시간에 따라 대여량의 흐름을 비교해 보았다.

### 각 시간대 별 Series 데이터 만들기
중간중간 빈 시간이 있어서 index를 사용한 시간 구분이 어렵다.

따라서 pandas 메소드를 사용해서 `datetime` 객체로 변환 후 진행했다

```python
data['dt_datetime'] = pd.to_datetime(data['datetime'],format='%Y-%m-%d %H:%M:%S')
```

`datetime` 객체는 각 날짜를 시간, 일, 월, 년, 요일 등으로 나누는 기능을 제공한다.

이를 활용해서 시간을 분류해서 계산했다.

```python
data["dt_hour"] = data["dt_datetime"].dt.hour
data["dt_day"] = data["dt_datetime"].dt.day
data["dt_year"] = data["dt_datetime"].dt.year
data["dt_month"] = data["dt_datetime"].dt.month
data["dt_dow"] = data["dt_datetime"].dt.dayofweek.map({0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday', 5:'Saturday',6:'Sunday'})
```

```python
plt.suptitle("시간당 그래프", size='xx-large')
sns.barplot(x = data["dt_hour"], y = data["count"])
plt.show()
```

새벽 시간에는 대여량이 적은 모습을 보였으며,
가장 대여량이 많은 시간대는 8시, 17시, 18시정도였다.

통근 시간대로 추정되며, 이 시간에 자전거를 대여해 통근하는 사람들이 많을 것으로 생각된다.

```python
f, axes = plt.subplots(ncols = 2, figsize=(10,5), sharey=True)
plt.suptitle("시간당 그래프", size='xx-large')
axes[0].set_title("비가입자")
axes[1].set_title("가입자")
sns.barplot(x = data["dt_hour"], y = data["casual"], ax = axes[0])
sns.barplot(x = data["dt_hour"], y = data["registered"], ax = axes[1])
plt.show()
```

비가입자는 통근시간대(8시, 17시, 18시)와 관련 없이 낮시간대 이용이 높은 모습을 보인다.

즉 낮에 이동이나 여가생활을 위해 사용하는 사람이 많을 것이라고 생각된다.

반면 가입자는 통근시간대 사용량이 크게 증가하는 것을 확인할 수 있다. 즉 이는 출퇴근을 위해 자전거를 정기적으로 이용하는 가입자가 많다는 것을 의미할 것이다

### 가입한 사용자는 정말로 출퇴근을 위해 사용하는가?

가입한 사용자가 통근을 위해 자전거를 사용하는 것이 맞는지에 대한 가설 검증을 해 보고자 한다.

자전거를 통근을 위해 사용하는 것이 맞다면, 평일 사용량이 주말에 비해 크게 나타날 것이고,

마찬가지로 이에 더해 평일에만 통근시간대에 사용량이 높게 나타난다면 가입한 사용자는 통근시간에 주로 자전거를 사용한다는 가설은 맞다고 생각할 수 있을 것이다.

`workingday` 를 기준으로 시간 별로 총 합을 구해서 비교해 보았다.

```python
hour_data_workday= data[data["workingday"]==1].groupby(data["dt_hour"]).mean()
hour_data_holiday= data[data["workingday"]==0].groupby(data["dt_hour"]).mean()
```

```python
f, axes = plt.subplots(nrows=2, ncols = 2, figsize=(10,5), sharey=True)
plt.suptitle("평일, 주말 시간당 그래프", size='xx-large')

axes[0, 0].set_title("평일 비가입자")
axes[0, 1].set_title("평일 가입자")
axes[1, 0].set_title("주말 비가입자")
axes[1, 1].set_title("주말 가입자")

sns.barplot(x = data["dt_hour"], y = hour_data_workday["casual"], ax = axes[0, 0])
sns.barplot(x = data["dt_hour"], y = hour_data_workday["registered"], ax = axes[0, 1])
sns.barplot(x = data["dt_hour"], y = hour_data_holiday["casual"], ax = axes[1, 0])
sns.barplot(x = data["dt_hour"], y = hour_data_holiday["registered"], ax = axes[1, 1])
plt.show()
```

예상했던 대로, 평일 가입자의 사용이 주를 이루고 있다.

또한 평일 가입자의 사용량은 통근 시간대에 높게 나타난것을 확인할 수 있다.

따라서 가입자는 평일 통근을 위해서 자전거를 사용한다고 생각할 수 있다.

```python
plt.suptitle("날짜별 그래프", size='xx-large')
sns.barplot(x=data["dt_day"], y = data["count"])
plt.show()
```

일별 차이는 크게 나타나지 않았다.

```python
plt.suptitle("월별 그래프", size='xx-large')
sns.barplot(x=data["dt_month"], y = data["count"])
plt.show()
```

겨울(1월, 2월)에 사용량이 적은 것을 확인할 수 있었다.
겨울철에는 날씨가 추워서 더 적은 사람들이 이용한 것으로 예상된다.

```python
plt.suptitle("요일별 그래프", size='xx-large')
sns.barplot(x=data["dt_dow"], y = data["count"])
plt.show()
```

요일 별 큰 차이는 없는 것을 확인할 수 있었다

## 결론

### 이상치 분석
- 데이터의 히스토그램을 그려 본 결과, 습도에서 0이 나오는 부분은 이상치로 추정되어 이전 시간의 수치로 대체했다.
- 풍속의 경우, 바람이 거의 안 부는 때는 측정이 안되는 것으로 의심이 되나, 수정하지 않았다.

### 상관관계 분석
- 주어진 데이터의 상관관계를 분석해 본 결과, 기온(체감온도)와 대여량은 양의 상관관계를 보인다.
- 습도와 대여량은 음의 상관관계를 보인다.
- `log_count` feature 를 만들어 비교해 본 결과, 습도와 대여량의 상관관계가 약간 강화되었으며, 기온과 대여량의 상관관계는 약간 약화되었다.
- `count`, `log_count` 모두에서 풍속과는 상관관계가 작게 나타났다.

### 계절별 분석 feature 분석
- 계절별로 데이터를 분석해 보았을 때, 총 대여량은 가을 > 여름 > 겨울 > 봄 순으로 유지되었다.
- 가입된 사용자는 여름보다 겨울에 사용하는 인원이 약간 많았으나, 비가입 사용자가 여름에 많이 유입되어 총 대여량이 여름에 더 많아진다.

### 평일, 휴일별 feature 분석
- 전체 사용자는 평일에 조금 더 많다.
- 비가입 사용자는 휴일에 자전거를 많이 대여하는 것을 확인할 수 있다. 여가 생활과 관련이 있는 것으로 생각이 된다.
- 비가입 사용자의 휴일 증가량보다 더 큰수로 가입된 사용자의 평일 사용량이 더 컸다. 통근 등을 위해 사용하는 것으로 추정된다.

### 날씨별 feature 분석
- 맑음, 흐림 등의 날씨인 경우(`weather == 0` 일 때) 사용량이 압도적으로 많았다.
- 예상대로 폭우(폭설)의 경우 (`weather == 3` 일 때) 사용량이 매우 적었다. 

### 날짜별 feature 분석
- 새벽 시간대는 예상 대로 사용률이 저조했다.
- 사용량이 가장 많은 시간은 8시, 17시, 18시인 출퇴근 시간이라 출퇴근 시간대에 사람이 많이 사용하는 것으로 예측이 가능했다.
- 비가입자의 경우, 출퇴근 사용자와는 반대로 14시 경이 가장 높은 사용량을 보였다. 낮시간 이동이나 여가를 위한 것으로 추측된다.
- 매월 1일부터 19일까지 사용량은 고루 분포된 모습을 확인했다.
- 월별로 비교해 본 결과 겨울(1월, 2월)의 사용량이 가장 적었고, 여름(6월)의 사용량이 가장 많았다.
- 요일별로 비교해 본 결과 큰 차이가 없었다.

### 가입자 비가입자 feature 분석
- 가입자의 경우는 평일 출퇴근시간대의 사용량이 매우 높았다.
- 비가입자의 경우는 평일 낮시간(15시~19시)의 사용량이 높았지만, 전반적으로 고르게 분포되어있었다.
- 또한 가입자와 비가입자 모두 주말은 낮 시간에 주로 사용한 것으로 나타났다. 평일과는 사용 목적이 다르다고 볼 수 있다.
- 따라서 가입자는 비가입자에 비해 평일 출퇴근 목적을 통해 주로 사용한다는 가설은 적절하다고 볼 수 있다..

### 연도별 feature 비교 분석
- 2011년에 비해 2012년의 자전거 대여량이 증가한 것을 확인할 수 있다.
- 월 별로 비교해 보았을 때, 대여량 추세에 큰 차이는 없다.
- 날짜와 시간으로도 비교해 볼 때, 큰 차이는 없다.
- 요일별 비교에서는 약간의 차이가 있었는데, 2011년에는 월요일, 금요일에 대여량이 높게 나타났지만 2012년에는 토요일에 대여량이 높게 나타났다
